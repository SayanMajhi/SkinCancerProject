# Data Preprocessing for HAM10000 Skin Cancer Dataset

In this notebook, we will handle the initial data preprocessing steps, including loading the dataset and performing basic exploration.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from PIL import Image

# Define paths
data_dir = 'C:\\Users\\Asus\\deeplearning-ham10000-workshop\\data'
images_dir = os.path.join(data_dir, 'images')
masks_dir = os.path.join(data_dir, 'masks')
csv_file = os.path.join(data_dir, 'GroundTruth.csv')

# Load the GroundTruth CSV file
ground_truth = pd.read_csv(csv_file)
ground_truth.head()
```

## Basic Exploration

Let's explore the dataset to understand the distribution of classes and the number of images available.

```python
# Count the number of images for each class
class_counts = ground_truth[['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']].sum()
class_counts.plot(kind='bar', title='Class Distribution')
plt.xlabel('Skin Cancer Types')
plt.ylabel('Number of Images')
plt.show()
```

## Load and Display Sample Images

Let's load and display a few sample images from the dataset to visualize the data.

```python
# Function to load and display images
def display_sample_images(image_ids, num_samples=6):
    plt.figure(figsize=(16, 6))
    for i, image_id in enumerate(image_ids[:num_samples]):
        img_path = os.path.join(images_dir, image_id + '.jpg')
        img = Image.open(img_path)
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(image_id)
    plt.show()

# Display sample images
sample_image_ids = ground_truth['image'].sample(6).tolist()
display_sample_images(sample_image_ids)
```
